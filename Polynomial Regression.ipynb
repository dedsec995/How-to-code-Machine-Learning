# Polynomial Regression

```python
#Data Preprocessing Template
#IMporting Libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Importing datasheet
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
#Fitting Linear Regression to the Dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
#fitting Polynomial Regression to the datast
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
#Visualing The linear Regression results
plt.scatter(X,y, color='red')
plt.plot(X,lin_reg.predict(X), color='blue')
plt.title("Truth or Bluff(Linear Regerssion)")
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

```python
#Visualing The Polynomial Regression results
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y, color='red')
plt.plot(X_grid,lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color='blue')
plt.title("Truth or Bluff(Linear Regerssion)")
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

```python
#Predicting a new result with Linear Regression
t = np.array([6.5])
t = t.reshape(-1,1)
lin_reg.predict(t)
```

```python
#Predicting a new result with Polynomial Regresion
t = np.array([6.5])
t = t.reshape(-1,1)
lin_reg_2.predict(poly_reg.fit_transform(t))
```
